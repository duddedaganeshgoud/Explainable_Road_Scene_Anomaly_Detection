```python
import os
import cv2
import numpy as np
import yaml
from tqdm import tqdm

# Set paths
BASE_PATH = '/kaggle/input/radroad-anomaly-detection'
YAML_PATH = '/kaggle/input/radroad-anomaly-detection/images/data.yaml'
OUTPUT_DIR = '/kaggle/working/organized_dataset'

# Load YAML configuration
def load_yaml_config():
    try:
        with open(YAML_PATH, 'r') as file:
            yaml_data = yaml.safe_load(file)
            print(f"Successfully loaded configuration from {YAML_PATH}")
            return yaml_data
    except Exception as e:
        print(f"Error loading YAML file: {e}")
        return None

# Create directory structure
def create_directory_structure(class_names):
    for split in ['train', 'valid', 'test']:
        for class_name in class_names:
            class_dir = os.path.join(OUTPUT_DIR, split, class_name)
            os.makedirs(class_dir, exist_ok=True)
    print(f"Created directory structure in {OUTPUT_DIR}")

# Process YOLO label file
def process_label_file(label_path, img_path, class_names, split):
    try:
        # Read image to get dimensions
        img = cv2.imread(img_path)
        if img is None:
            return
        
        height, width = img.shape[:2]
        
        # Read label file
        with open(label_path, 'r') as f:
            lines = f.readlines()
            
        # Process each object in the label file
        for line in lines:
            parts = line.strip().split()
            if len(parts) >= 5:
                class_id = int(float(parts[0]))
                x_center = float(parts[1])
                y_center = float(parts[2])
                w = float(parts[3])
                h = float(parts[4])
                
                # Skip if class_id is invalid
                if class_id >= len(class_names):
                    continue
                
                # Convert normalized coordinates to pixel values
                x1 = int((x_center - w/2) * width)
                y1 = int((y_center - h/2) * height)
                x2 = int((x_center + w/2) * width)
                y2 = int((y_center + h/2) * height)
                
                # Ensure coordinates are within image bounds
                x1 = max(0, x1)
                y1 = max(0, y1)
                x2 = min(width, x2)
                y2 = min(height, y2)
                
                # Skip if object is too small
                if x2 - x1 < 10 or y2 - y1 < 10:
                    continue
                
                # Extract the object
                obj_img = img[y1:y2, x1:x2]
                
                # Resize to standard size (224x224)
                obj_img = cv2.resize(obj_img, (224, 224))
                
                # Save to appropriate class directory
                class_name = class_names[class_id]
                output_filename = f"{os.path.splitext(os.path.basename(img_path))[0]}_obj{x1}_{y1}.jpg"
                output_path = os.path.join(OUTPUT_DIR, split, class_name, output_filename)
                
                # Check if class directory has 500 images already
                class_dir = os.path.join(OUTPUT_DIR, split, class_name)
                if len(os.listdir(class_dir)) >= 500:
                    continue
                    
                cv2.imwrite(output_path, obj_img)
    except Exception as e:
        print(f"Error processing {label_path}: {e}")

# Get all files in a directory with specific extension
def get_files_with_extension(directory, extension):
    return [f for f in os.listdir(directory) if f.endswith(extension)]

# Main function
def organize_dataset():
    # Load YAML config
    yaml_config = load_yaml_config()
    if not yaml_config:
        return
    
    # Get class names
    class_names = yaml_config.get('names', [])
    print(f"Found {len(class_names)} classes: {class_names}")
    
    # Create directory structure
    create_directory_structure(class_names)
    
    # Fix paths - use the correct directories
    # For Kaggle dataset structure, usually images are in /images and labels in /labels
    splits = {
        'train': 'train',
        'valid': 'valid',
        'test': 'test'
    }
    
    # Process each split
    for split_name, split_dir in splits.items():
        # Construct paths for images and labels
        img_dir = os.path.join(BASE_PATH, 'images', split_dir)
        label_dir = os.path.join(BASE_PATH, 'labels', split_dir)
        
        # Check if directories exist
        if not os.path.exists(img_dir):
            print(f"Image directory not found: {img_dir}")
            continue
        
        if not os.path.exists(label_dir):
            print(f"Label directory not found: {label_dir}")
            continue
        
        print(f"\nProcessing {split_name} split:")
        print(f"Images: {img_dir}")
        print(f"Labels: {label_dir}")
        
        # Get all label files
        label_files = get_files_with_extension(label_dir, '.txt')
        print(f"Found {len(label_files)} label files")
        
        # Process each label file
        for label_file in tqdm(label_files, desc=f"Processing {split_name}"):
            label_path = os.path.join(label_dir, label_file)
            
            # Find corresponding image file (try both jpg and png)
            img_base = os.path.splitext(label_file)[0]
            jpg_path = os.path.join(img_dir, img_base + '.jpg')
            png_path = os.path.join(img_dir, img_base + '.png')
            
            if os.path.exists(jpg_path):
                process_label_file(label_path, jpg_path, class_names, split_name)
            elif os.path.exists(png_path):
                process_label_file(label_path, png_path, class_names, split_name)
    
    # Print statistics
    print("\nDataset organization complete!")
    print("\nClass distribution:")
    for split in ['train', 'valid', 'test']:
        print(f"\n{split.capitalize()} set:")
        for class_name in class_names:
            class_dir = os.path.join(OUTPUT_DIR, split, class_name)
            if os.path.exists(class_dir):
                count = len(os.listdir(class_dir))
                print(f"  {class_name}: {count} images")
            else:
                print(f"  {class_name}: 0 images")

# Let's try with recursive search in case the directory structure is different
def find_label_files(base_path):
    """Recursively search for label files"""
    print("Searching for label files recursively...")
    
    label_files = []
    for root, dirs, files in os.walk(base_path):
        for file in files:
            if file.endswith('.txt') and 'classes.txt' not in file and 'data.yaml' not in file:
                label_files.append(os.path.join(root, file))
    
    print(f"Found {len(label_files)} label files")
    
    # Try to understand the directory structure
    label_dirs = set()
    for label_file in label_files:
        label_dirs.add(os.path.dirname(label_file))
    
    print("Label directories found:")
    for label_dir in label_dirs:
        print(f"  {label_dir}")
    
    return label_files, label_dirs

def organize_dataset_recursive():
    # Load YAML config
    yaml_config = load_yaml_config()
    if not yaml_config:
        return
    
    # Get class names
    class_names = yaml_config.get('names', [])
    print(f"Found {len(class_names)} classes: {class_names}")
    
    # Create directory structure
    create_directory_structure(class_names)
    
    # Find all label files
    label_files, label_dirs = find_label_files(BASE_PATH)
    
    # Determine the splits based on directory names
    split_mapping = {}
    for label_dir in label_dirs:
        if 'train' in label_dir.lower():
            split_mapping[label_dir] = 'train'
        elif 'valid' in label_dir.lower() or 'val' in label_dir.lower():
            split_mapping[label_dir] = 'valid'
        elif 'test' in label_dir.lower():
            split_mapping[label_dir] = 'test'
        else:
            # Default to train if can't determine
            split_mapping[label_dir] = 'train'
    
    print("\nAssigned splits:")
    for label_dir, split in split_mapping.items():
        print(f"  {label_dir} -> {split}")
    
    # Process each label file
    processed_count = 0
    for label_path in tqdm(label_files, desc="Processing labels"):
        label_dir = os.path.dirname(label_path)
        split = split_mapping.get(label_dir, 'train')  # Default to train
        
        # Find corresponding image file
        img_path = None
        img_dir = label_dir.replace('labels', 'images')
        
        if not os.path.exists(img_dir):
            # Try other common patterns
            img_dir = os.path.join(os.path.dirname(os.path.dirname(label_dir)), 'images', os.path.basename(label_dir))
            if not os.path.exists(img_dir):
                continue
        
        img_base = os.path.splitext(os.path.basename(label_path))[0]
        jpg_path = os.path.join(img_dir, img_base + '.jpg')
        png_path = os.path.join(img_dir, img_base + '.png')
        
        if os.path.exists(jpg_path):
            img_path = jpg_path
        elif os.path.exists(png_path):
            img_path = png_path
        else:
            continue
        
        # Process this label and image
        process_label_file(label_path, img_path, class_names, split)
        processed_count += 1
    
    print(f"\nProcessed {processed_count} label files")
    
    # Print statistics
    print("\nDataset organization complete!")
    print("\nClass distribution:")
    for split in ['train', 'valid', 'test']:
        print(f"\n{split.capitalize()} set:")
        for class_name in class_names:
            class_dir = os.path.join(OUTPUT_DIR, split, class_name)
            if os.path.exists(class_dir):
                count = len(os.listdir(class_dir))
                print(f"  {class_name}: {count} images")
            else:
                print(f"  {class_name}: 0 images")

if __name__ == "__main__":
    # Try the regular approach first
    organize_dataset()
    
    # Check if we got any data
    any_data = False
    for split in ['train', 'valid', 'test']:
        for class_name in load_yaml_config().get('names', []):
            class_dir = os.path.join(OUTPUT_DIR, split, class_name)
            if os.path.exists(class_dir) and len(os.listdir(class_dir)) > 0:
                any_data = True
                break
    
    # If no data, try the recursive approach
    if not any_data:
        print("\nNo data found using standard paths. Trying recursive search...")
        organize_dataset_recursive()
```

```python
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

# Set path to organized dataset
DATASET_PATH = '/kaggle/working/organized_dataset'

# Set constants
IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 16
EPOCHS = 20
CLASS_NAMES = ['HMV', 'LMV', 'Pedestrian', 'RoadDamages', 'SpeedBump', 'UnsurfacedRoad']
NUM_CLASSES = len(CLASS_NAMES)

# Enable GPU memory growth to avoid OOM errors
physical_devices = tf.config.list_physical_devices('GPU')
if len(physical_devices) > 0:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)
    print("GPU memory growth enabled")

# Load images from a directory
def load_images_from_directory(directory):
    images = []
    image_files = [f for f in os.listdir(directory) if f.endswith(('.jpg', '.jpeg', '.png'))]
    
    for img_file in tqdm(image_files, desc=f"Loading {os.path.basename(directory)}"):
        img_path = os.path.join(directory, img_file)
        try:
            img = cv2.imread(img_path)
            if img is None:
                continue
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
            img = img / 255.0  # Normalize
            images.append(img)
        except Exception as e:
            print(f"Error loading {img_path}: {e}")
    
    return np.array(images)

# Load dataset from organized directory structure
def load_dataset(split):
    X = []
    y = []
    
    split_dir = os.path.join(DATASET_PATH, split)
    if not os.path.exists(split_dir):
        print(f"Directory not found: {split_dir}")
        return np.array([]), np.array([])
    
    print(f"\nLoading {split} dataset:")
    
    for class_idx, class_name in enumerate(CLASS_NAMES):
        class_dir = os.path.join(split_dir, class_name)
        if not os.path.exists(class_dir):
            print(f"Class directory not found: {class_dir}")
            continue
        
        # Load images from this class
        class_images = load_images_from_directory(class_dir)
        
        if len(class_images) > 0:
            X.append(class_images)
            y.append(np.full(len(class_images), class_idx))
            print(f"  {class_name}: {len(class_images)} images")
    
    if X:
        X = np.vstack(X)
        y = np.concatenate(y)
        print(f"Total {split} images: {len(X)}")
        return X, y
    else:
        return np.array([]), np.array([])

# Build CNN model
def build_model():
    model = Sequential([
        # First convolution block
        Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.25),
        
        # Second convolution block
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.25),
        
        # Third convolution block
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.25),
        
        # Fourth convolution block
        Conv2D(256, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(0.25),
        
        # Fully connected layers
        Flatten(),
        Dense(512, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(256, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(NUM_CLASSES, activation='softmax')
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Train and evaluate model
def train_and_evaluate():
    # Load datasets
    X_train, y_train = load_dataset('train')
    X_val, y_val = load_dataset('valid')
    X_test, y_test = load_dataset('test')
    
    if len(X_train) == 0:
        print("No training data found. Exiting.")
        return
    
    # Build model
    print("\nBuilding CNN model...")
    model = build_model()
    model.summary()
    
    # Prepare callbacks
    callbacks = [
        ModelCheckpoint('/kaggle/working/best_model.keras', monitor='val_accuracy', save_best_only=True, mode='max'),
        EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
        ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)
    ]
    
    # Convert labels to one-hot encoding
    y_train_one_hot = to_categorical(y_train, num_classes=NUM_CLASSES)
    
    # Validation data if available
    validation_data = None
    if len(X_val) > 0:
        y_val_one_hot = to_categorical(y_val, num_classes=NUM_CLASSES)
        validation_data = (X_val, y_val_one_hot)
    
    # Train model
    print("\nTraining model...")
    history = model.fit(
        X_train, y_train_one_hot,
        batch_size=BATCH_SIZE,
        epochs=EPOCHS,
        validation_data=validation_data,
        callbacks=callbacks
    )
    
    # Save final model
    model.save('/kaggle/working/final_model.keras')
    print("Model saved to /kaggle/working/final_model.h5")
    
    # Plot training history
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    if validation_data:
        plt.plot(history.history['val_accuracy'])
        plt.legend(['Train', 'Validation'])
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    if validation_data:
        plt.plot(history.history['val_loss'])
        plt.legend(['Train', 'Validation'])
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/training_history.png')
    
    # Evaluate on test set
    if len(X_test) > 0:
        print("\nEvaluating on test set...")
        y_test_one_hot = to_categorical(y_test, num_classes=NUM_CLASSES)
        model.evaluate(X_test, y_test_one_hot)
        
        # Generate predictions
        y_pred = np.argmax(model.predict(X_test), axis=1)
        
        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='weighted')
        recall = recall_score(y_test, y_pred, average='weighted')
        f1 = f1_score(y_test, y_pred, average='weighted')
        
        print(f"Test Accuracy: {accuracy:.4f}")
        print(f"Test Precision: {precision:.4f}")
        print(f"Test Recall: {recall:.4f}")
        print(f"Test F1 Score: {f1:.4f}")
        
        # Classification report
        print("\nClassification Report:")
        print(classification_report(y_test, y_pred, target_names=CLASS_NAMES))
        
        # Confusion matrix
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
        plt.title('Confusion Matrix')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.tight_layout()
        plt.savefig('/kaggle/working/confusion_matrix.png')
    else:
        print("No test data available for evaluation")

if __name__ == "__main__":
    train_and_evaluate()
```
